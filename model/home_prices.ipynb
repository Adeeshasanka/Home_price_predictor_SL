Data Cleaning

```python
#suppose these features has no impact on final price
df2 = df1.drop(['Title','Sub_title','Address','Description','Post_URL','Seller_name','Seller_type','published_date','Geo_Address','Lat','Lon'], axis=1)
df2.head()
```

```python
df2.isnull().sum()
```

```python
df2.Beds.unique()
```

```python
df2[df2.Beds=='10+']
```

```python
df3 = df2.copy()
df3['Bedrooms'] = df2['Beds'].apply(lambda x: 10 if x == '10+' else x)
df3.loc[15013]
```

```python
df3.Baths.unique()
```

```python
df3['Bathrooms'] = df2['Baths'].apply(lambda x: 10 if x =='10+' else int(x))
df3.loc[9235]
```

```python
df4 = df3.drop(['Beds', 'Baths'], axis=1)
df4.head(3)
```

```python
df4['Land_perch'] = df3['Land size'].apply(lambda x: float(x.split(' ')[0].replace(',','')))
df4.head(3)
```

```python
df4['size_sqft'] = df3['House size'].apply(lambda x: float(x.split(' ')[0].replace(',','')))
df4.head(3)
```

```python
df5 = df4.drop(['House size', 'Land size'], axis=1)
df5.head(3)
```

```python
df5['Town'] = df4['Location'].apply(lambda x: x.split(', ')[0])
df5['District'] = df4['Location'].apply(lambda x: x.split(', ')[1])
df5.head(5)
```

```python
df5['Price_Rs'] = df4['Price'].apply(lambda x: int(x.split(' ')[1].replace(',','')))
df5.head(3)
```

```python
df5.groupby('Town')['Town'].agg('count')
```

```python
df5.District.unique()
```

```python
df6 = df5.drop(['Location','Price'], axis=1)
df6.head()
```

```python
df6['Price_per_sqft'] = df5['Price_Rs']/df5['size_sqft']
df6.head(3)
```

```python
len(df6.Town.unique())
```

```python
df6.Bedrooms = pd.to_numeric(df6['Bedrooms'], errors='coerce')
```

Outlier detection and Removal

```python
len(df6[df6.size_sqft/df6.Bedrooms < 200])
```

```python
df7 = df6[~(df6.size_sqft/df6.Bedrooms < 200)]
df7.shape
```

```python
df7.Price_per_sqft.describe()
```

```python
def remove_pps_outliers(df):
    df_out = pd.DataFrame()
    for key, subdf in df.groupby('District'):
        m = np.mean(subdf.Price_per_sqft)
        st = np.std(subdf.Price_per_sqft)
        reduced_df = subdf[(subdf.Price_per_sqft>(m-st) & subdf.Price_per_sqft<=(m+st))]
        df_out = pd.concat([df_out,reduced_df],ignore_index=True)
    return df_out

df8 = remove_pps_outliers(df7)
df7.shape
```
