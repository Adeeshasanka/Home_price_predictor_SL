Outlier detection and Removal

```python
len(df6[df6.size_sqft/df6.Bedrooms < 200])
```

```python
df7 = df6[~(df6.size_sqft/df6.Bedrooms < 200)]
df7.shape
```

```python
df7.Price_per_sqft.describe()
```

```python
def remove_pps_outliers(df):
    df_out = pd.DataFrame()
    for key, subdf in df.groupby('District'):
        mean = np.mean(subdf.Price_per_sqft)
        st_deviation = np.std(subdf.Price_per_sqft)
        reduced_df = subdf[(subdf.Price_per_sqft>(mean-st_deviation)) & (subdf.Price_per_sqft<=(mean+st_deviation))]
        df_out = pd.concat([df_out,reduced_df],ignore_index=True)
    return df_out

df8 = remove_pps_outliers(df7)
df8.shape
```

```python
df8['District'] = df8['District'].str.strip()
df8['Town'] = df8['Town'].str.strip()
df8.head()
```

```python
# Function to plot a scatter chart comparing 2-bedroom and 3-bedroom properties in a specific town
def plot_chart_2vs3_bed(df,Town):
    bed2 = df[(df.Town == Town) & (df.Bedrooms == 2)]  # Filter data for 2-bedroom properties in the specified town
    bed3 = df[(df.Town == Town) & (df.Bedrooms == 3)]  # Filter data for 3-bedroom properties in the specified town
    matplotlib.rcParams['figure.figsize'] = (4,3)
    # Plot a scatter chart for 2-bedroom properties with blue color and label
    plt.scatter(bed2.size_sqft, bed2.Price_Rs, color='blue', label='2 bed', s=50)
    # Plot a scatter chart for 3-bedroom properties with green color and '+' marker
    plt.scatter(bed3.size_sqft, bed3.Price_Rs, color='green',marker='+', label='3 bed', s=50)
    plt.xlabel('Size_sqft')
    plt.ylabel('price')
    plt.title(Town)
    plt.legend()
    plt.show()

plot_chart_2vs3_bed(df8, 'Malabe')
```

```python
def remove_bedroom_outliers(df):
    # Initialize an empty array to store the indices of outliers
    exclude_indices = np.array([])
    # Group the DataFrame by 'Town' to process each town separately
    for town, town_df in df.groupby('Town'):
        # Dictionary to store mean, standard deviation, and count for each bedroom category
        bedroom_stats = {}
        # Group the town-specific DataFrame by 'Bedrooms' to calculate stats for each bedroom type
        for bedrooms, bedroom_df in town_df.groupby('Bedrooms'):
            bedroom_stats[bedrooms] = {
                'mean': np.mean(bedroom_df.Price_per_sqft), # Calculate the mean price per square foot
                'std': np.std(bedroom_df.Price_per_sqft),   # Calculate the standard deviation
                'count': bedroom_df.shape[0]                # Count the number of data points
            }
        for bedrooms, bedroom_df in town_df.groupby('Bedrooms'):
            # Get the stats of one-bedroom category lower than the current one
            stats = bedroom_stats.get(bedrooms-1)
            # Proceed only if the previous bedroom category exists and has more than 5 data points
            if stats and stats['count']>5:
                 # Append the indices of properties that are priced lower than the mean of the smaller bedroom category
                exclude_indices = np.append(exclude_indices, bedroom_df[bedroom_df.Price_per_sqft<(stats['mean'])].index.values)
    return df.drop(exclude_indices, axis='index')

# Apply the function to remove outliers from the dataset
df9 = remove_bedroom_outliers(df8)
df9.shape
```

Model Building

```python
df_dummies = pd.get_dummies(df11[['Town', 'District']],dtype=int, drop_first=True)
df_dummies.head()
```

```python
df_dummies.shape
```

```python
len(df11.District.unique())
```

```python
df12 = pd.concat([df11.drop(['District', 'Town'], axis=1),df_dummies], axis=1)
df12.head()
```

```python
x = df12.drop(['Price_Rs'], axis=1)
y = df12.Price_Rs
x.head()
```

```python
y.head()
```

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2, random_state=10)

x_train.shape
```

```python
from sklearn.model_selection import ShuffleSplit
from sklearn.model_selection import cross_val_score

cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)
```

```python
from sklearn.linear_model import LinearRegression

scroes = cross_val_score(LinearRegression(), x, y, cv=cv)
np.mean(scroes)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.linear_model import Lasso
from sklearn.tree import DecisionTreeRegressor

def find_the_best_model(x,y):
    models = {
        'linear_regression': {
            'model': LinearRegression(),
            'params': {
                'fit_intercept': [True, False]
            }
        },
        'lasso': {
            'model': Lasso(),
            'params': {
                'alpha': [1,2],
                'selection': ['random', 'cyclic']
            }
        },
        'decision_tree': {
            'model': DecisionTreeRegressor(),
            'params': {
                'criterion': ['mse', 'friedman_mse'],
                'splitter': ['best', 'random']
            }
        }
    }
    scores = []
    cv = ShuffleSplit(n_splits=5, test_size=0.2, random_state=0)
    for model, config in models.items():
        gs = GridSearchCV(config['model'], config['params'], return_train_score=False, cv=cv)
        gs.fit(x,y)
        scores.append({
            'model': model,
            'best_params': gs.best_params_,
            'best_score': gs.best_score_
        }) 
    return pd.DataFrame(scores)
```

```python
find_the_best_model(x,y)
```

```python
#choosing DecisionTreeRegressor as the ideal model basing on the above data
model = DecisionTreeRegressor(criterion='friedman_mse', splitter='random')
model.fit(x_train, y_train)
model.score(x_test,y_test)
```

```python
def predict_price(Town, District, sqft, bedrooms,bathrooms, land_perch, x):
    if ('Town_'+Town) in x.columns:
        loc_index_Town = np.where(x.columns == 'Town_'+Town)[0][0]
    else:
        loc_index_Town = None
    if ('District_'+District) in x.columns:
        loc_index_District = np.where(x.columns == 'District_'+District)[0][0]
    else:
        loc_index_District = None

    x = np.zeros(len(x.columns))
    x[0] = bedrooms
    x[1] = bathrooms
    x[2] = land_perch
    x[3] = sqft
    if loc_index_District:
        x[loc_index_District] = 1
    if loc_index_Town:
        x[loc_index_Town] = 1
    return model.predict([x])[0]
```
